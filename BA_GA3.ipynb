```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
train_df = pd.read_csv("train_LR.csv")
test_df = pd.read_csv("test_LR.csv")
```

```python
Q1 = train_df.shape[0]
Q1
```

```python
Q2 = test_df.shape[0]
Q2
```

```python
X_train = train_df.drop("Accept", axis=1)
y_train = train_df["Accept"]

X_test = test_df.drop("Accept", axis=1)
y_test = test_df["Accept"]
```

```python
Q3 = ( (y_train == 0).mean() * 100 )
round(Q3, 4)
```

```python
Q4 = ( (y_train == 1).mean() * 100 )
round(Q4, 4)
```

```python
Q5 = ( (y_test == 0).mean() * 100 )
round(Q5, 4)
```

```python
Q6 = ( (y_test == 1).mean() * 100 )
round(Q6, 4)
```

```python
model = LogisticRegression(random_state=17, max_iter=1000)
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```

```python
Q7 = accuracy_score(y_train, y_pred_train) * 100
round(Q7, 4)
```

```python
Q8 = precision_score(y_train, y_pred_train, pos_label=0) * 100
round(Q8, 4)
```

```python
Q9 = recall_score(y_train, y_pred_train, pos_label=1) * 100
round(Q9)
```

```python
cm = confusion_matrix(y_test, y_pred_test)
tn, fp, fn, tp = cm.ravel()
cm
```

```python
train_prec_1 = precision_score(y_train, y_pred_train, pos_label=1)
test_prec_1  = precision_score(y_test,  y_pred_test,  pos_label=1)

if test_prec_1 > train_prec_1:
    Q15 = "Increased"
elif test_prec_1 < train_prec_1:
    Q15 = "Decreased"
else:
    Q15 = "Both are equal"

Q15
```
